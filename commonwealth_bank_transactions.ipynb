```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('supermarket_transactions.xlsx')
df
```

```python
df = df.drop(labels='Unnamed: 0', axis=1)
```

```python
df.info()
```

```python
apples_df = df[['product_name', 'quantity','store', 'payment_method','customer_id']]
apples_df = apples_df.set_index(keys='store')
apples_df
```

```python
apples_df = apples_df[(apples_df['product_name'] == 'apple') & (apples_df['payment_method'] == 'cash')]
apples_df = apples_df.reset_index()
apples_df
```

```python
apples_df['quantity'].sum()
```

```python
cash_df = apples_df
cash_df = cash_df.merge(df[['total_amount','customer_id']], how="left", on='customer_id')
cash_df
```

```python
cash_df['total_amount'].sum()
```

```python
bakershire_expenditure_df = df[['customer_type', 'payment_method','store','total_amount']]
bakershire_expenditure_df = bakershire_expenditure_df[(bakershire_expenditure_df['customer_type'] == 'non-member') & (bakershire_expenditure_df['store'] == 'Bakershire')]
bakershire_expenditure_df = bakershire_expenditure_df.reset_index(drop=True)
bakershire_expenditure_df
```

```python
bakershire_expenditure_df['total_amount'].sum()
```

```python
with pd.ExcelWriter("analysis.xlsx", engine="xlsxwriter") as writer:
    df.to_excel(writer, sheet_name="Raw_Data", index=False)
    apples_df.to_excel(writer, sheet_name="apples_purchased_in_cash_data", index=False)
    cash_df.to_excel(writer, sheet_name="total_cash_spendings_on_apples", index=False)
    bakershire_expenditure_df.to_excel(writer, sheet_name="money_spent_at_bakershire_data", index=False)
```
